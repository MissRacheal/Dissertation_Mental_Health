```python
import pandas as pd

# Raw data provided
data = {
    'Age Group': ['All adults', '16 to 39', '40 to 69', '70 and over'],
    'Percentage': [16, 23, 15, 8],
    'Lower Bound': [15, 19, 13, 6],
    'Upper Bound': [18, 26, 17, 9]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print(df)
```

```python
import pandas as pd

from matplotlib import pyplot as plt
%matplotlib inline

import seaborn as sns

print("Matplotlib and Seaborn are imported")
```

```python
# Bar plot for percentage by age group
plt.figure(figsize=(10, 6))
sns.barplot(x='Age Group', y='Percentage', data=df)
plt.title('Percentage of Mental Health Issues by Age Group')
plt.ylabel('Percentage')
plt.xlabel('Age Group')
plt.show()

# Error bars to show lower and upper bounds
plt.figure(figsize=(10, 6))
plt.errorbar(df['Age Group'], df['Percentage'], yerr=[df['Percentage'] - df['Lower Bound'], df['Upper Bound'] - df['Percentage']], fmt='o')
plt.title('Mental Health Issues by Age Group with Confidence Intervals')
plt.ylabel('Percentage')
plt.xlabel('Age Group')
plt.show()
```
